```python
import numpy as np
import matplotlib.pyplot as plt
from split_comparison_experiment import *
%run constants.py
```

```python
algs = ['svm', 'lr', 'nn']
# data to plot
n_groups = 3
decentr = [alg_svm_aver_train_time_decentr, alg_lr_aver_train_time_decentr, alg_nn_aver_train_time_decentr]
baseline = [alg_svm_aver_train_time_baseline, alg_lr_aver_train_time_baseline, alg_nn_aver_train_time_baseline]
cent = [alg_svm_aver_train_time_real, alg_lr_aver_train_time_real, alg_nn_aver_train_time_real]
# create plot
fig, ax = plt.subplots(figsize=(10, 6))
index = np.arange(n_groups)
bar_width = 0.4
opacity = 0.8
ax.set_ylim(ymin=0.0)
ax.set_ylim(ymax=0.1)
rects1 = plt.bar(index, decentr, bar_width,
                 alpha=opacity,
                 color='b',
                 label='Decentralized')
rects1 = plt.bar(index + bar_width, baseline, bar_width,
                 alpha=opacity,
                 color='g',
                 label='Baseline')

plt.xlabel('Algorithm')
plt.ylabel('Time')
plt.title('Training time for different algorithms (5000000 samples, 1000 models, 5 models per group)')
plt.xticks(index + bar_width/2, algs)
plt.legend()

fig, ax = plt.subplots(figsize=(9, 6))
index = np.arange(n_groups)
bar_width = 0.4
opacity = 0.8
ax.set_ylim(ymin=10)
ax.set_ylim(ymax=1200)
rects1 = plt.bar(index, centr, bar_width,
                 alpha=opacity,
                 color='r',
                 label='Centralized')
 
plt.xlabel('Algorithm')
plt.ylabel('Time')
plt.title('Training time for different algorithms (5000000 samples, 1000 models, 5 models per group)')
plt.xticks(index, algs)
plt.legend()
 
 
plt.tight_layout()
plt.show()
```

```python
# simple split 
algs = ['svm', 'lr', 'nn']
# data to plot
n_groups = 3
thund = [alg_svm_aver_simple, alg_lr_aver_simple, alg_lr_aver_simple]
baseline = [alg_svm_aver_simple_baseline, alg_lr_aver_simple_baseline, alg_nn_aver_simple_baseline]
centr = [alg_svm_aver_real, alg_lr_aver_real, alg_nn_aver_real]

# create plot
fig, ax = plt.subplots(figsize=(12, 10))
index = np.arange(n_groups)
bar_width = 0.2
opacity = 0.8
ax.set_ylim(ymin=0.7)
ax.set_ylim(ymax=1.0)
rects1 = plt.bar(index, thund, bar_width,
                 alpha=opacity,
                 color='b',
                 label='Thunderdome')
 
rects2 = plt.bar(index + bar_width, baseline, bar_width,
                 alpha=opacity,
                 color='g',
                 label='Baseline')

rects2 = plt.bar(index + 2*bar_width, centr, bar_width,
                 alpha=opacity,
                 color='r',
                 label='Centralized')
 
plt.xlabel('Algorithm')
plt.ylabel('AUC')
plt.title('AUC for SIMPLE SPLIT (5000000 samples, 1000 models, 5 models per group)')
plt.xticks(index + bar_width, algs)
plt.legend()
 
plt.tight_layout()
plt.show()
```
