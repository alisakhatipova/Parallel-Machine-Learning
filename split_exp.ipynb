```python
import numpy as np
import matplotlib.pyplot as plt
from split_comparison_experiment import *
%run constants.py
```

```python
algs = ['svm', 'logistic regression', 'multilayer perceptron']
# data to plot
n_groups = 3
decentr = [alg_svm_aver_train_time_decentr, alg_lr_aver_train_time_decentr, alg_nn_aver_train_time_decentr]
baseline = [alg_svm_aver_train_time_baseline, alg_lr_aver_train_time_baseline, alg_nn_aver_train_time_baseline]
centr = [alg_svm_aver_train_time_real, alg_lr_aver_train_time_real, alg_nn_aver_train_time_real]
# create plot
fig, ax = plt.subplots(figsize=(12, 10))
index = np.arange(n_groups)
bar_width = 0.2
opacity = 0.8
rects1 = plt.bar(index, decentr, bar_width,
                 alpha=opacity,
                 color='b',
                 label='Main')
rects1 = plt.bar(index + bar_width, baseline, bar_width,
                 alpha=opacity,
                 color='g',
                 label='Baseline')
rects2 = plt.bar(index + 2*bar_width, centr, bar_width,
                 alpha=opacity,
                 color='r',
                 label='Centralized')
plt.xlabel('Algorithm', fontsize = 'xx-large')
plt.ylabel('Log(Time)', fontsize = 'xx-large')
plt.title('Time measurements (500000 samples, 1000 learners, 5 learners per group)', fontsize = 'xx-large')
plt.xticks(index + bar_width/2, algs, fontsize = 'xx-large')
plt.yticks(fontsize = 'xx-large')
plt.legend(fontsize = 'xx-large', loc = 9)
ax.set_yscale('log')

plt.tight_layout()
plt.show()
```

```python
# simple split 
algs = ['svm', 'logistic regression', 'multilayer perceptron']
# data to plot
n_groups = 3
thund = [alg_svm_aver_simple, alg_lr_aver_simple, alg_lr_aver_simple]
baseline = [alg_svm_aver_simple_baseline, alg_lr_aver_simple_baseline, alg_nn_aver_simple_baseline]
centr = [alg_svm_aver_real, alg_lr_aver_real, alg_nn_aver_real]

# create plot
fig, ax = plt.subplots(figsize=(12, 10))
index = np.arange(n_groups)
bar_width = 0.2
opacity = 0.8
ax.set_ylim(ymin=0.775)
ax.set_ylim(ymax=0.795)
rects1 = plt.bar(index, thund, bar_width,
                 alpha=opacity,
                 color='b',
                 label='Main')
 
rects2 = plt.bar(index + bar_width, baseline, bar_width,
                 alpha=opacity,
                 color='g',
                 label='Baseline')

rects2 = plt.bar(index + 2*bar_width, centr, bar_width,
                 alpha=opacity,
                 color='r',
                 label='Centralized')
 
plt.xlabel('Algorithm', fontsize = 'xx-large')
plt.ylabel('AUC', fontsize = 'xx-large')
plt.title('AUC for SIMPLE SPLIT (5000000 samples, 1000 learners, 5 learners per group)', fontsize = 'xx-large')
plt.xticks(index + bar_width, algs, fontsize = 'xx-large')
plt.yticks(fontsize = 'xx-large')
plt.legend(fontsize = 'xx-large')
 
plt.tight_layout()
plt.show()
```
