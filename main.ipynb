```python
import numpy as np
import matplotlib.pyplot as plt
%run constants.py
```

```python
algs = ['svm', 'lr', 'nn']
# data to plot
n_groups = 3
thund = [0.82, 0.8100, 0.85]
baseline = [0.827, 0.8116, 0.85]
centr = [0.9, 0.96, 0.94]

# create plot
fig, ax = plt.subplots(figsize=(12, 10))
index = np.arange(n_groups)
bar_width = 0.2
opacity = 0.8
ax.set_ylim(ymin=0.7)
rects1 = plt.bar(index, thund, bar_width,
                 alpha=opacity,
                 color='b',
                 label='Thunderdome', yerr=[0.1, 0.1, 0.1])
 
rects2 = plt.bar(index + bar_width, baseline, bar_width,
                 alpha=opacity,
                 color='g',
                 label='Baseline')

rects2 = plt.bar(index + 2*bar_width, centr, bar_width,
                 alpha=opacity,
                 color='r',
                 label='Centralized')
 
plt.xlabel('Algorithm')
plt.ylabel('AUC')
plt.title('AUC for different algorithms (30 models)')
plt.xticks(index + bar_width, algs)
plt.legend()
 
plt.tight_layout()
plt.show()
```
