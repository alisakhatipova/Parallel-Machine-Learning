```python
import numpy as np
import matplotlib.pyplot as plt
from group_size_comparison_experiment import *
%run constants.py
```

```python
# group size 2
algs = ['svm', 'lr', 'nn']
# data to plot
n_groups = 3
thund = [alg_svm2_aver_normal, alg_lr2_aver_normal, alg_nn2_aver_normal]
baseline = [alg_svm2_aver_baseline, alg_lr2_aver_baseline, alg_nn2_aver_baseline]
centr = [alg_svm2_aver_real, alg_lr2_aver_real, alg_nn2_aver_real]

# create plot
fig, ax = plt.subplots(figsize=(12, 10))
index = np.arange(n_groups)
bar_width = 0.2
opacity = 0.8
ax.set_ylim(ymin=0.7)
rects1 = plt.bar(index, thund, bar_width,
                 alpha=opacity,
                 color='b',
                 label='Thunderdome')
 
rects2 = plt.bar(index + bar_width, baseline, bar_width,
                 alpha=opacity,
                 color='g',
                 label='Baseline')

rects2 = plt.bar(index + 2*bar_width, centr, bar_width,
                 alpha=opacity,
                 color='r',
                 label='Centralized')
 
plt.xlabel('Algorithm')
plt.ylabel('AUC')
plt.title('AUC for different algorithms (30 models)')
plt.xticks(index + bar_width, algs)
plt.legend()
 
plt.tight_layout()
plt.show()
```
