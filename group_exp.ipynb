```python
import numpy as np
import matplotlib.pyplot as plt
from group_size_comparison_experiment import *
%run constants.py
```

```python
# group size 2
algs = ['svm', 'logistic regression', 'multilayer perceptron']
# data to plot
n_groups = 3
thund = [alg_svm2_aver_normal, alg_lr2_aver_normal, alg_nn2_aver_normal]
baseline = [alg_svm2_aver_baseline, alg_lr2_aver_baseline, alg_nn2_aver_baseline]
centr = [alg_svm2_aver_real, alg_lr2_aver_real, alg_nn2_aver_real]

# create plot
fig, ax = plt.subplots(figsize=(12, 10))
index = np.arange(n_groups)
bar_width = 0.2
opacity = 0.8
ax.set_ylim(ymin=0.775)
ax.set_ylim(ymax=0.795)
rects1 = plt.bar(index, thund, bar_width,
                 alpha=opacity,
                 color='b',
                 label='Main')
 
rects2 = plt.bar(index + bar_width, baseline, bar_width,
                 alpha=opacity,
                 color='g',
                 label='Baseline')

rects2 = plt.bar(index + 2*bar_width, centr, bar_width,
                 alpha=opacity,
                 color='r',
                 label='Centralized')
 
plt.xlabel('Algorithm', fontsize = 'xx-large')
plt.ylabel('AUC', fontsize = 'xx-large')
plt.title('AUC for GROUP SIZE 2 (5000000 samples, 1000 learners)', fontsize = 'xx-large')
plt.xticks(index + bar_width, algs, fontsize = 'xx-large')
plt.yticks(fontsize = 'xx-large')
plt.legend(fontsize = 'xx-large')
 
plt.tight_layout()
plt.show()
```
